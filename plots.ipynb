```python
import pandas as pd
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming cars_data is already defined and loaded
x = cars_data['KM']
y = cars_data['Weight']
z = cars_data['Price']

# Create 3D plot
ax = plt.axes(projection='3d')
ax.plot_trisurf(x, y, z, cmap='jet')

# Set title
ax.set_title('3D Surface Plot')

# Display the plot
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cars_data = pd.read_csv('./Toyota.csv')
cars_data.dropna(axis=0,inplace=True)
```

```python
#scatter plot
sns.set(style='darkgrid')
sns.regplot(x=cars_data['Age'],y=cars_data['Price'])
```

```python
#boxplot
plt.boxplot([cars_data['Price'],cars_data['HP'],cars_data['KM']])
plt.xticks([1,2,3],['price','HP','KM'])
```

```python
#heatmaps
sns.heatmap(cars_data[['Price','Age','KM','HP']].corr(),cmap='jet')
plt.show()
```

```python
#counterplot
x=cars_data['KM']
y=cars_data['Weight']
z=cars_data['Price']
plt.tricontourf(x,y,z,level=20,cmap='jet')
plt.colorbar(label='Price')
plt.xlabel('KM')
plt.ylabel('Weight')
plt.title('counter_plot')
plt.show()
```

```python
#3d surface plot
from mpl_toolkits.mplot3d import Axes3D
x=cars_data['KM']
y=cars_data['Doors']
z=cars_data['Price']
ax=plt.axes(projection='3d')
ax.plot_trisurf(x,y,z,cmap='jet')
ax.set_title('3d plot')
plt.show()
```
